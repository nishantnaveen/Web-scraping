# Scraping Headline,published time and corresponding News link



```python
import requests
import bs4
from bs4 import BeautifulSoup as bs
import pandas as pd

page= requests.get('https://www.cnbc.com/world/?region=world%20:')
# page

soup=bs(page.content)
# soup

headline=[]
time=[]
newslink=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)

for i in soup.find_all('span',class_="LatestNews-wrapper"):
    time.append(i.text)

for i in soup.find_all("a", class_="LatestNews-headline"):
    newslink.append(i.get('href'))

df=pd.DataFrame({'Headline':headline,'Published Time': time,'URL': newslink})
df
```
