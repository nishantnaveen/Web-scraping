# Scraping top 10 batsmen in men's ODI

```python
import requests
from bs4 import BeautifulSoup as bs
page= requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
# page

soup=bs(page.content)
# soup

ranking=[]
player_name=[]
team=[]
rating=[]

# scraping the rank 1 ODI batsman

rank1= soup.find('div',class_="rankings-block__top-player")
ranking.append(rank1.text.split('\n')[4].strip(' '))
player_name.append(rank1.text.split('\n')[13])
team.append(rank1.text.split('\n')[16])
rating.append(rank1.text.split('\n')[17].strip(' '))

# print(ranking,player_name,team,rating)

# df=pd.DataFrame({'POSITION': position, 'PLAYER':player_name, 'TEAM': team,'RATING': rating})
# df[:10]

# # problem : 7,8 and 9th rank did not populate

# 7 8 and 9 did not populate because the ranking of these 3 players changed. the change in ranking is shown
# just after the position, thus changing some attribute in its HTML source file. need to find that and then extract those 3 
# player detail

for i in soup.find_all('tr',class_="table-body"):
    ranking.append(i.text.split('\n')[4].strip(' '))
    if i.text.split('\n')[8]== '(0)':
        player_name.append(i.text.split('\n')[13])
        team.append(i.text.split('\n')[17])
        rating.append(i.text.split('\n')[19])
    else:
        player_name.append(i.text.split('\n')[16])
        team.append(i.text.split('\n')[20])
        rating.append(i.text.split('\n')[22])
        
    
import pandas as pd    
df=pd.DataFrame({'POSITION': ranking, 'PLAYER':player_name, 'TEAM': team,'RATING': rating})
df[:10]

```
