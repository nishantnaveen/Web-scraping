# Scraping top 10 bowlers data from ICC webpage

```python
import requests
import bs4
from bs4 import BeautifulSoup as bs

page=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')

# page

soup=bs(page.content)
# soup

ranking=[]
player=[]
team=[]
rating=[]

bowler1=player1 = soup.find('tr', class_="rankings-block__banner")
ranking.append(bowler1.text.split('\n')[4].strip(' '))
player.append(bowler1.text.split('\n')[20])
team.append(bowler1.text.split('\n')[27])
rating.append(bowler1.text.split('\n')[31])

for i in soup.find_all('tr', class_="table-body"):
    ranking.append(i.text.split('\n')[4].strip(' '))
    if i.text.split('\n')[8]=='(0)':
        player.append(i.text.split('\n')[13])
        team.append(i.text.split('\n')[17])
        rating.append(i.text.split('\n')[19])
    else:
        player.append(i.text.split('\n')[16])
        team.append(i.text.split('\n')[20])
        rating.append(i.text.split('\n')[22])
        
import pandas as pd
df=pd.DataFrame({'Ranking': ranking,'Player':player,'Team':team,'Rating':rating})
df[:10]
```
