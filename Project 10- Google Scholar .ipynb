# Scraping the Rank,publication,h5-index and h5 median of top publications from Google Scholar form.

```python
# importing all the required libraries

# importing requests library which helps in pulling out contents of a website
import requests 
import bs4
from bs4 import BeautifulSoup as bs
import pandas as pd

page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
# page

content= bs(page.content)
# content

# getting the HTML content using bs4/beautifulsoup


rank=[]
publication=[]
h5_index=[]
h5_median=[]

for i in content.find_all('td', class_="gsc_mvt_p"):
    rank.append(i.text)  

for i in content.find_all('td', class_="gsc_mvt_t"):
    publication.append(i.text)

for i in content.find_all('a', class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(i.text)

for i in content.find_all('span', class_="gs_ibl gsc_mp_anchor"):
    h5_median.append(i.text)
    
# print(h5_index)

df= pd.DataFrame({'Ranking':rank, 'Publication':publication, 'h5-index': h5_index[:100], 'h5-median':h5_median})
df
```
