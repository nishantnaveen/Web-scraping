# Scraping top 100 IMDB indian movies

```python
import requests   
import bs4 
from bs4 import BeautifulSoup as bs

link= requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
link
soup=bs(link.content)

movie_name=[]
year=[]


for i in soup.find_all('td', class_="titleColumn"):
    movie_name.append(i.text.split('\n')[2])
    year.append(i.text.split('\n')[3])
    
# # year[:100]

ratings=[]
for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    ratings.append(i.text.replace('\n', ""))
# ratings[:100]

# Importing Pandas
import pandas as pd
df=pd.DataFrame({'Movies':movie_name, 'Year of Release': year, 'Ratings': ratings})
df[:100]
```
