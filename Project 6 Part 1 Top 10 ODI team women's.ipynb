# Scraping top 10 ODI women's team

```python
import requests
import bs4
from bs4 import BeautifulSoup as bs
import pandas as pd

page=requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
# page

soup=bs(page.content)
# soup

position=[]
team=[]
matches=[]
points=[]
rating=[]

topteam=soup.find('tr',class_="rankings-block__banner")
position.append(topteam.text.split('\n')[1])
team.append(topteam.text.split('\n')[4])
matches.append(topteam.text.split('\n')[7])
points.append(topteam.text.split('\n')[8])
rating.append(topteam.text.split('\n')[10].strip(' '))

# print(position,team,matches,points,rating)

for i in soup.find_all('tr',class_="table-body"):
    position.append(i.text.split('\n')[1])
    team.append(i.text.split('\n')[4])
    matches.append(i.text.split('\n')[7])
    points.append(i.text.split('\n')[8])
    rating.append(i.text.split('\n')[9])




df=pd.DataFrame({'POSITION': position[:10],'TEAM':team[:10],'MATCHES':matches[:10],'POINTS':points[:10],'RATINGS':rating[:10]})
df
```
