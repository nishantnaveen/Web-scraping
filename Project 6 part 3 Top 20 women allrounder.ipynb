```python
import requests
import bs4
from bs4 import BeautifulSoup as bs

page=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder')

# page
soup=bs(page.content)
# soup

ranking=[]
player=[]
team=[]
rating=[]

# scraping data for the top player since it has a different class/attribute than the rest of the table

player1 = soup.find('tr', class_="rankings-block__banner")
ranking.append(player1.text.split('\n')[4].strip(' '))
player.append(player1.text.split('\n')[20])
team.append(player1.text.split('\n')[27])
rating.append(player1.text.split('\n')[31])

# Scraping data for the rest 9 players

for i in soup.find_all('tr', class_="table-body"):
    ranking.append(i.text.split('\n')[4].strip(' '))
    player.append(i.text.split('\n')[13])
    team.append(i.text.split('\n')[17])
    rating.append(i.text.split('\n')[19])
    
        
import pandas as pd
df=pd.DataFrame({'Ranking': ranking,'Player':player,'Team':team,'Rating':rating})
df[:10]
```
