# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
b) Top 10 womenâ€™s ODI Batting players along with the records of their team and rating.
    

```python
# importing all the required libraries

# importing requests library which helps in pulling out contents of a website
import requests 
import bs4
from bs4 import BeautifulSoup
import pandas as pd

page = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi')

# getting the HTML content using bs4/beautifulsoup
soup= BeautifulSoup(page.content)
# soup

# creating empty list for the data 
ranking=[]
player=[]
team=[]
ratings=[]

# scraping first woman batsman individually as this is in a banner format hence cannot be scrapped with rest of the players
player1 = soup.find('div', class_="rankings-block__top-player")
ranking.append(player1.text.split('\n')[4].strip(' '))
player.append(player1.text.split('\n')[13])
team.append(player1.text.split('\n')[16])
ratings.append(player1.text.split('\n')[17].strip(' '))

# print (ranking, player, team, ratings)

# scraping rest of the teams together as they are in tabular format

for i in soup.find_all('tr',class_="table-body"):
    ranking.append(i.text.split('\n')[4].strip(' '))
    player.append(i.text.split('\n')[13])
    team.append(i.text.split('\n')[17])
    ratings.append(i.text.split('\n')[19])
             

df=pd.DataFrame({'ODI Ranking':ranking, 'Woman Batsmen':player, 'Team':team, 'Ratings':ratings})
df[:10]
```
